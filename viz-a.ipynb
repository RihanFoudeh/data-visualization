```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## gitting the datasets from seaborn


```python
sns.set()
```

## Load the data receive into a Pandas DataFrame

```python
df = sns.load_dataset('penguins')
df=pd.DataFrame(df)
df
```

## Analyzing the penguins dataset

```python
sns.barplot(x='bill_length_mm', y='body_mass_g', data=df.head(15))
```

```python
sns.lmplot(x='bill_length_mm', y='body_mass_g', data=df)
```

```python
sns.relplot(x='bill_length_mm', y='body_mass_g', data=df)
```
